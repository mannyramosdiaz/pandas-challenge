## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#creating db for top spender group
top_spender_group=purchase_data.groupby(["SN"])
top_spender_group.first()
```

```python
#Top spender count, counts items purchased
top_spender_count=round(top_spender_group["Purchase ID"].count(),2)
top_spender_count
#mean or average of the average spent per item
top_spender_mean=round(top_spender_group["Price"].mean(),2)
top_spender_mean
#total spent by user
top_spender_ave=round(top_spender_count*top_spender_mean,2)
top_spender_ave
```

```python
#table with count of items, average of amount per item and total spent by members in descending order
top_spender_df=pd.DataFrame({"Purchase Count":top_spender_count, "Average Purchase Price":top_spender_mean,"Total Purchase Value":top_spender_ave})
top_spender_df=top_spender_df.sort_values("Total Purchase Value",ascending=False)
top_spender_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#making of the db to hold the data, grouping by item in order of their id number
most_popular_group=purchase_data.groupby(["Item ID","Item Name"])
most_popular_group.first()
```

```python
#counting the amount of times an id is purchased
most_popular_count=round(most_popular_group["Purchase ID"].count(),2)
most_popular_count
#getting the average price of items 
most_popular_mean=round(most_popular_group["Price"].mean(),2)
most_popular_mean
#using both to get the total amount of money spent per item average
most_popular_ave=round(most_popular_count*most_popular_mean,2)
most_popular_ave
```

```python
#Making a table with the parameters
most_popular_df=pd.DataFrame({"Purchase Count":most_popular_count, "Average Purchase Price":most_popular_mean,"Total Purchase Value":most_popular_ave})
most_popular_df=most_popular_df.sort_values("Purchase Count",ascending=False)
most_popular_df
```
